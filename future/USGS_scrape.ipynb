## 1. iterated through all of the NOAA prediction pages
    pull link on the far right to USGS gauge

```python
USGS_links = []
for g in df['NOAA_gauge']:
    link = 'https://www.cbrfc.noaa.gov/station/flowplot/flowplot.cgi?' + g
    response = requests.get(link)
    soup = BeautifulSoup(response.text, "html.parser")
    div_L1 = soup.find_all('div')
    div_L2 = div_L1[1]
    div_L3 = div_L2.find_all('div')
    # some div_L3 don't have any links, skip those
    value = 0
    if len(div_L3) > 0:
        a1 = div_L3[0].find_all('a')
        for a in a1:
            if a.get_text() == 'USGS data':
                value = a['href']
    USGS_links.append(value)
    time.sleep(3)
```

```python
# save these while I run the loop again; this was 346 entries on the previous try through the loop
USGS_links
```

```python
len(USGS_links)
```

## add the USGS links to the main DF

```python
df['USGS_link'] = USGS_links
df
```

```python
df['usgs'] = df['USGS_link'].str.split('?')
```

```python
df.tail()
```

```python
df['usgs']
```

## Extract USGS number and add to DF as another column

```python
usgs_g = []
for us in df['usgs']:
    if type(us) != float:
        usgs_g.append(us[1])
    else:
        usgs_g.append(0)
usgs_g
```

```python
df['usgs'] = usgs_g
df
```
